```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def getkb(X,Y):
    k=(np.average(Y*X)-np.average(Y)*np.average(X))/(np.average(X*X)-np.average(X)**2)
    b=np.average(Y)-k*np.average(X)
    def av(n):
        return np.average(n)
    sigk=np.sqrt((av(Y*Y)-av(Y)*av(Y))/(av(X*X)-av(X)**2) - k*k)/np.sqrt(X.shape[0])
    sigb=sigk*np.sqrt(av(X*X)-av(X)**2)
    return k,b,sigk,sigb
```

```python
izm=np.array(range(40,180,20))
izm
```

```python
cl=[]
for i in izm:
    cl.append(0)
    k=0
    for d in open(str(i)+'.txt'):
        cl[-1]+=int(d)
        k+=1
    cl[-1]/=k
    
cl=np.array(cl)
```

```python
kp,bp,_,_=getkb(cl,izm)
```

```python
plt.grid(which='major')
plt.grid(which='minor', linestyle=':')
plt.minorticks_on()

plt.xlabel('Давление, мм рт. ст.')
plt.ylabel(r'Отсчёты АЦП')
plt.title('График калибровки')

plt.plot(cl*kp+bp,cl,'black',linewidth=0.5)
plt.scatter(izm,cl,50,'r',marker='x',linewidths=1)
```

```python
name='ac0p'
dat=pd.read_csv(name+'.txt', header=None, names=['p','t'])
print(dat)
dat['p']=kp*dat['p']+bp
```

```python
plt.scatter(dat['t'],dat['p'],0.1)
```

```python
from scipy.signal import savgol_filter
T=np.array(dat['t'])
P=np.array(dat['p'])
P=P[np.where((1.8<T)&(T<70))[0]]
T=T[np.where((1.8<T)&(T<70))[0]]
#P= savgol_filter(P, 1000, 1)
```

```python
plt.plot(T,P)
```

```python
k=200
def filt(X,Y,k):
    Yn=np.zeros(Y.shape[0]-2*k)
    for i in range(k,Y.shape[0]-k):
        Yn[i-k]=np.sum(Y[i-k:i+k])/(2*k)
    Xn=X[k:-k]
    return Xn,Yn
Tn,Pn=filt(T,P,k)
```
